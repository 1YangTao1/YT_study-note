```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_housing_data():
    csv_path = "./datasets/housing/housing.csv"
    return pd.read_csv(csv_path)
```

```python
h_info = load_housing_data()
h_info.info()
```

```python
h_info.head()
```

```python
h_info['ocean_proximity'].value_counts()
```

```python
h_info.describe()
```

```python
h_info.hist(bins = 50,figsize = (20,15))
#bins条形数，figszie创建图的大小
```

```python
h_info.plot(kind = 'scatter',x = 'longitude',y='latitude',alpha = 0.4,s = h_info['population']/100,label = "population",figsize = (10,7),
           c = 'median_house_value',cmap = plt.get_cmap('jet'),colorbar = True)
plt.legend()
```
